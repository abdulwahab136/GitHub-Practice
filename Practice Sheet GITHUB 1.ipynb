Q5: A football is kicked 1 meter from the ground at an angle of 30 degree with a velocity of 15 m/sec. 
<br>Calculate the following:
<br>a. Max height above the ground.
<br>b. Total time in the air.
<br>c. Total distance traveled.
<br>d. Velocity vector at max height.
<br>e. Velocity when the ball hits the ground.

![proj](proj.jpg)



```python
from math import sin,cos,sqrt

velocity = 15
gforce = 9.8
angle = 30

#max height above ground
maxh = (velocity**2*sin(angle)**2)/ (2*gforce)
print("maximam height above the ground is ",maxh,"meters")

#total time in the air
T = (2*velocity*sin(angle))/(-gforce)
print("total time football was in Air is ",T,"seconds")

#Total distance travelled 
totald =-(velocity*T*sin(angle)-1/2*gforce*T**2)
print("Total distance covered by football : ",totald,"meters")

#velocity vector at max height
vvector = (velocity*sin(angle))-(-gforce*T)
print("velocity vector at max height is ",vvector,"m/s")

#velocity when ball hit ground 
time = sqrt(2*maxh/gforce)
vground = velocity+gforce*time
print("velocity of object when it hit the ground is " ,vground,"m/s")


print("\n\n time tag : 20+ minutes")
```
